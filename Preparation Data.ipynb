```python
# wpp = xr.open_dataset('/content/drive/MyDrive/overfishnt/data/weather_and_marine_condition/swh/koordinat.txt')

file_path = '/content/drive/MyDrive/overfishnt/Machine_Learning/data/weather_and_marine_condition/swh/koordinatwpp.xlsx'  # Replace with the actual file path
# Read the Excel file into a Pandas DataFrame
wpp = pd.read_excel(file_path)
```

```python

#2022
prp22 = xr.open_dataset('/content/drive/MyDrive/overfishnt/Machine_Learning/data/weather_and_marine_condition/presipitasi/presipitasi_2022.nc')
prp22 = prp22.to_dataframe()
prp22= prp22.reset_index()

#2023
prp23 = xr.open_dataset('/content/drive/MyDrive/overfishnt/Machine_Learning/data/weather_and_marine_condition/')
presipitasi/presipitasi_2023.nc')
prp23 = prp23.to_dataframe()
prp23 = prp23.reset_index()
prp23


```

```python
list_wpp = ['571','572','573','711','712','713','714',
            '715','716','717','718']
for i in range (len(wpp)):
  header = wpp.iloc[i]
  x = prp22.loc[(prp22['longitude']==header['longitude']) & 
   (prp22['latitude']==header['latitude'])]
  y = prp23.loc[(prp23['longitude']==header['longitude']) & 
   (prp23['latitude']==header['latitude'])]
  new_df = pd.concat([x, y])
  nama = list_wpp[i]
  new_df.to_csv('/content/drive/MyDrive/overfishnt/data/weather_and_marine_condition/presipitasi/csv/'+f'prep{nama}.csv',index=False)
  f"prep{nama}.csv"
  # print(header)
```
