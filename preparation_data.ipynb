```python
import warnings
warnings.filterwarnings('ignore')

import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def extract_file_nc(param, wpp):
  ds = xr.open_dataset('/content/drive/MyDrive/overfishnt/data/fishing_spot_heatmap/'+param+'/'+wpp+'.nc')
  if param == 'chl':
    pass
  else:
    ds = ds.resample(time='D').mean()
  df = ds.to_dataframe().sort_values(by=['latitude','longitude'])
  df2=df.reset_index()
  for i in range(0, len(ds.longitude)):
    for j in range(0, len(ds.latitude)):
      lon = ds.longitude[i].values
      lat = ds.latitude[j].values
      new_df = df2.loc[(df2['longitude']==lon) & (df2['latitude']==lat)]
      count_nan = new_df[param].isna().sum()
      if count_nan == 0:
        break
  new_df = new_df.dropna().reset_index(drop=True)
  new_df = new_df[['time', param]]
  print(lon,lat,wpp)
  new_df.to_csv('/content/drive/MyDrive/overfishnt/data/fishing_spot_heatmap/'+param+'/csv_file/'+wpp+'.csv',index=False)
  
```

```python
WPP = ['WPP571', 'WPP572', 'WPP573', 'WPP711', 'WPP712', 'WPP713', 
       'WPP714', 'WPP715', 'WPP716', 'WPP717', 'WPP718']
param = ['chl','o2','sst']
```

```python
for i in WPP:
  for j in param : 
    extract_file_nc(j, i)
```
